```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt

class Expert(nn.Module):
    def __init__(self, input_dim, hidden_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 1)
        )
    
    def forward(self, x):
        return self.net(x)

class GatingNetwork(nn.Module):
    def __init__(self, input_dim, num_experts):
        super().__init__()
        self.fc = nn.Linear(input_dim, num_experts)
    
    def forward(self, x):
        return F.softmax(self.fc(x), dim=-1)

class SparseMoE(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_experts, k=1):
        super().__init__()
        self.experts = nn.ModuleList([Expert(input_dim, hidden_dim) for _ in range(num_experts)])
        self.gate = GatingNetwork(input_dim, num_experts)
        self.k = k
        self.num_experts = num_experts
    
    def forward(self, x):
        gate_weights = self.gate(x)
        topk_weights, topk_indices = torch.topk(gate_weights, k=self.k, dim=-1)  

        sparse_weights = torch.zeros_like(gate_weights).scatter_(
            dim=-1, index=topk_indices, src=topk_weights
        )  

        if self.training:
            expert_mask = F.one_hot(topk_indices, num_classes=self.num_experts).float()  
            selection_prob = expert_mask.mean(dim=0) 
            load_balancing_loss = self._load_balancing_loss(gate_weights, selection_prob)
            self.loss = load_balancing_loss

        expert_outputs = torch.stack([expert(x) for expert in self.experts], dim=1)
        output = torch.sum(sparse_weights.unsqueeze(-1) * expert_outputs, dim=1) 
        return output
    
    def _load_balancing_loss(self, gate_weights, selection_prob):
        prob_per_expert = gate_weights.mean(dim=0)
        cov_per_expert = selection_prob.mean(dim=0)  
        balance_loss = torch.sum(prob_per_expert * cov_per_expert) * self.num_experts
        return balance_loss
```

```python
def create_data(n=1000):
    x = np.random.uniform(-3, 3, n)
    y = x ** 2 + np.random.normal(0, 1, n)
    return torch.tensor(x, dtype=torch.float32).unsqueeze(1), torch.tensor(y, dtype=torch.float32).unsqueeze(1)

X_train, y_train = create_data()
model = SparseMoE(input_dim=1, hidden_dim=16, num_experts=4, k=2)  # 激活Top-2专家
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
loss_fn = nn.MSELoss()
```

```python
for epoch in range(3000):
    model.train()
    pred = model(X_train)
    loss = loss_fn(pred, y_train)
    
    # 添加负载均衡损失（需加权调节）
    total_loss = loss + 0.01 * model.loss
    
    optimizer.zero_grad()
    total_loss.backward()
    optimizer.step()
    
    if epoch % 500 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}, Balance Loss: {model.loss.item():.4f}")
```

```python
model.eval()
x_test = torch.linspace(-3, 3, 200).unsqueeze(1)
y_pred = model(x_test).detach().numpy()

plt.figure(figsize=(8, 4))
plt.scatter(X_train.numpy(), y_train.numpy(), s=5, alpha=0.3, label='Train Data')
plt.plot(x_test.numpy(), y_pred, color='red', label='Sparse MoE (Top-2)')
plt.legend()
plt.title("Sparse MoE Regression: y = x^2")
plt.show()
```
